```python
a = int(input('Nhập giá trị của a: '))
b = int(input('Nhập giá trị của b: '))

import matplotlib.pyplot as plt
x = list(range(-5, 5))

y = []  

for i in x:
    formular = a * i + b
    y.append(formular)

plt.plot(x, y, color="r", marker="o")
plt.title("Phương trình y = ax + b", fontsize=16, fontweight="bold", color="blue")
plt.xlabel("Giá trị của x", fontsize=14, color="green")
plt.ylabel("Giá trị của y", fontsize=14, color="red")
plt.xticks(x)
plt.yticks(y)
plt.show()
```

```python
import csv
Date_list = []
HighFixed_list = []
with open(file_path, newline='', encoding='utf-8') as csv_file:
    data_csv = csv.reader(csv_file, delimiter=',')
    header = next(data_csv)
    for row in data_csv:
        Date_list.append((row[1]))
        HighFixed_list.append((row[3]))

import matplotlib.pyplot as plt

plt.plot(Date_list[:10], HighFixed_list[:10], color = 'b', marker = 'v') 
plt.title("Biểu đồ chứng khoán", fontsize=14, color="blue")
plt.xlabel("Date", fontsize=14, color="green")
plt.ylabel("HighFixed", fontsize=14, color="red")
plt.xticks(rotation=45) 
plt.show()
```

```python
a = int(input('Nhập giá trị của a: '))
b = int(input('Nhập giá trị của b: '))
c = int(input('Nhập giá trị của c: '))

import matplotlib.pyplot as plt
x = list(range(-4, 6))

y = []  

for i in x:
    formular = (a * i)**2 + b * i + c
    y.append(formular)

plt.plot(x, y, color="r", marker="o")
plt.title("Phương trình y = ax^2 + bx + c", fontsize=16, fontweight="bold", color="blue")
plt.xlabel("Giá trị của x", fontsize=14, color="green")
plt.ylabel("Giá trị của y", fontsize=14, color="red")
plt.xticks(x)
plt.yticks(y)
plt.show()
```
