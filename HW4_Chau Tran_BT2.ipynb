```python
import pandas as pd
import numpy as np
import matplotlib 
from matplotlib import pyplot as plt

file_path = r"C:\Users\HCM-LA129\Desktop\DA10\Week 3\sale_data_vn.csv"

df = pd.read_csv(file_path)
print(type(df))
```

```python
df.head()
```

```python
City_counts = df['City'].value_counts()
Gender_counts = df['Gender'].value_counts()
print (City_counts)
print (Gender_counts)
```

```python
fig = plt.figure(figsize=(8,6), constrained_layout=True)
ax1 = fig.add_subplot(1, 2, 1)

ax1.bar(Gender_counts.index, Gender_counts, color='skyblue', width = 0.5)
ax1.set_title('Biểu đồ tần số đơn hàng theo giới tính')
       
ax1.set_xlabel('Giới tính')
ax1.set_ylabel('Tần số đặt hàng')

ax1 = fig.add_subplot(1, 2, 2)
ax1.bar(City_counts.index, City_counts, color='r', width = 0.75)
ax1.set_title('Biểu đồ tần số đơn hàng theo thành phố')
       
ax1.set_xlabel('Thành phố')
ax1.set_ylabel('Tần số đặt hàng')

plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
ax1 = fig.add_subplot(1, 2, 1)
ax1.pie(City_counts, labels = City_counts.index, autopct='%.1f%%')
ax1.set_title("Biểu đồ tần suất đơn hàng theo thành phố", fontsize=10, color="y")

ax2 = fig.add_subplot(1, 2, 2)
ax2.pie(Gender_counts, labels = Gender_counts.index, autopct='%.1f%%')
ax2.set_title("Biểu đồ tần suất đơn hàng theo giới tính", fontsize=10, color="y")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 8))

ax1.hist(df['Total'], bins=25, edgecolor='black', alpha=0.7)
ax1.set_xlabel('Tổng giá trị')
ax1.set_ylabel('Tần số')
ax1.set_title('Biểu đồ phân bố giá trị đơn hàng')

ax2.hist(df['Rating'], bins=5, edgecolor='black', alpha=0.7, rwidth=2)
ax2.set_xlabel('Rating')
ax2.set_ylabel('Tần số')
ax2.set_title('Biểu đồ phân bố giá trị Ratings')

plt.tight_layout()
plt.show()
```
