```python
# Lấy ra thông tin từ các cột Diện tích, Chiều dài, Chiều rộng, Giá nhà
new_df = df[["Diện tích", "Dài", "Rộng", "Giá/m2"]]
print(new_df)
```

```python
# Tạo bảng dữ liệu copy
df_copy = df.copy()
df_copy.head()
```

```python
# Xóa cột Giấy tờ pháp lý và Số tầng

df_copy = df_copy.drop(columns = ["Giấy tờ pháp lý", "Số tầng"])
df_copy
```

```python
# Lấy đơn vị khỏi các cột Dài, Rộng

df_copy['Dài'] = df_copy['Dài'].str.rstrip('m')
df_copy['Rộng'] = df_copy['Rộng'].str.rstrip('m')

df_copy['Dài'] = df_copy['Dài'].astype(float)
df_copy['Rộng'] = df_copy['Rộng'].astype(float)
df_copy
```

```python
# Split cột Diện tích và Giá/m2

new_area = df_copy['Diện tích'].str.split(" ", n=1, expand = True)
df_copy['Diện tích'] = new_area[0]
df_copy['Bỏ'] = new_area[1]
df_copy = df_copy.drop(columns = ['Bỏ'])

new_price = df_copy["Giá/m2"].str.split(" ", n = 1, expand = True)
df_copy["Giá/m2"] = new_price[0]
df_copy["Bỏ"] = new_price[1]
df_copy = df_copy.drop(columns = ["Bỏ"])

# Xóa các dấu "." và thay dấu "," bằng dấu ".". Do số của Việt Nam sử dụng "." cho hàng nghìn và "," cho thập phân

df_copy["Giá/m2"] = df_copy["Giá/m2"].str.replace(".", "")
df_copy["Giá/m2"] = df_copy["Giá/m2"].str.replace(",", ".")

df_copy
```

```python
# Chuyển đổi cột "Ngày" thành kiểu dữ liệu datetime
df_copy['Ngày'] = pd.to_datetime(df_copy['Ngày'])

df_copy['Ngày'] = df_copy['Ngày'].dt.strftime('%d/%m/%Y')
df_copy
```

```python
# Lưu DataFrame vào tệp CSV mới
df_copy.to_csv('VN_housing_dataset_cleaned.csv', index=False)
```
